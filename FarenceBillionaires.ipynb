## 4) How many male billionaires are there compared to the number of female billionares? What percent is that? Do they have a different average wealth?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
femaleB = df[df['gender'].str.contains('female', na=False)].count()
femaleB['gender']

totalB = df[df['gender'].str.contains('male', na=False)].count()
totalB['gender']

print(f"There are {femaleB['gender']} female billionaires in this data set.")
print(f"There are {totalB['gender'] - femaleB['gender']} male billionaires in this data set.")

maleB = totalB['gender'] - femaleB['gender']
int(maleB)


percent_male = round((float(maleB/totalB['gender']) * 100),2)


print(f"{percent_male:,}% of billionaires in this list are male.")
```

```python
df.groupby(by='gender').median().sort_values(by='gender',ascending=False)
```
